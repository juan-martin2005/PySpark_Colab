# Iniciar sesión de Spark


```python
from pyspark.sql import SparkSession

# Crear una sesión de Spark
spark = SparkSession.builder \
    .appName("Consulta CSV con Spark") \
    .getOrCreate()
sc = spark.sparkContext
```

```python
df = spark.read.csv("supermarket_sales.csv", header=True, inferSchema=True)
df.show()
```

```python
# Lectura del archivo CSV (https://www.kaggle.com/datasets/aungpyaeap/supermarket-sales)
rdd = df.rdd

rdd.collect()
```

# Operaciones RDD - Conjunto de Datos Distribuidos

# Tranformaciones

```python
#map
# Aplicamos una funcion para obtener el Nombre del Producto su Precio unitario y la Fecha
rdd.map(lambda row: (row['Product line'], row['Unit price'], row['Date'])).collect()
```

```python
#filter
# Aplicamos un filtro por una fecha en especifico
rdd_filter = rdd.filter(lambda row: row['Date'] == '1/2/2019')
for i in rdd_filter.collect():
  print(i)
```

```python
#flatMap
# El flatmap funciona igual al MAP con la unica diferencia que "aplana" los valores es decir los pone en una sola lista
rdd.flatMap(lambda row: [(row['Product line'], row['Unit price'], row['Date'])]).collect()
```

```python
#union

# Filtramos a los clientes para despues unirlos y obtener la cantidad por separado
rdd_m = rdd.filter(lambda row: row['Gender'] == 'Male')
rdd_f = rdd.filter(lambda row: row['Gender'] == 'Female')

union_rdd = rdd_m.union(rdd_f)

# Seleccionamos solo los datos de la columna Género
gender = union_rdd.map(lambda row: [row['Gender']]).collect()

print(gender)
```

```python
#distinc - Nos devuelve valores unicos esto nos ayuda a evitar la redundancia de datos ala hora de realizar consultas

rdd_product = rdd.map(lambda row: row['Product line']).distinct()

# Mostramos todos los Productos / Linea de Productos de nuestor CSV
rdd_product.collect()
```

```python
#groupByKey -  Realiza el agrupamiento por clave de un conjunto de datos
rdd_product_quantity = rdd.map(lambda row:(row['Product line'], row['Quantity']))
group = rdd_product_quantity.groupByKey().mapValues(list).collect()

print(group)
```

```python
#reduceByKey - Devuelve un conjunto de datos de pares usando la funcion de reducción

rdd_product_quantity = rdd.map(lambda row:(row['Product line'], row['Quantity']))
group = rdd_product_quantity.reduceByKey(lambda x,y:x+y).collect()
print(group)
```

```python
#sortByKey - Esta función ordena los datos del RDD de entrada
rdd_product_sort = rdd.map(lambda row: row['Product line']).distinct()
group = rdd_product_sort.sortByKey(True)
group.collect()
```

# Acciones

```python
# reduce
fecha = "1/1/2019"
fecha_filter_rdd = rdd.filter(lambda row: row['Date'] == fecha)
# Usamos reduce para sumar el total de alguna fecha
total_rdd = fecha_filter_rdd.map(lambda row: float(row['Total'])).reduce(lambda x, y: x + y)

print(f"Total de ventas en la fecha {fecha}: ${total_rdd}")
```

```python
#colletc - convierte el RDD en una coleccion de array y lo muestra
rdd.collect()
```

```python
#count - devuelve el número de elemenetos del dataset
rrd_cantidad = rdd.count()
print(f"Cantidad de registros: {rrd_cantidad}")
```

```python
#first - devuelve el primer elemento del conjunto de datos

primer_elemento = rdd.first()
print(f"Primer elemento :\n{primer_elemento}")
```

```python
#take - devuelve un array con los primeros n elementos del dataset
rdd.take(5)
```

```python
#saveAsTextFile - Guarda el RDD como un archivos de texto en el directorio que quieras
rdd.saveAsTextFile("supermarket_sales.txt")
```

```python
#max, min

max_total = rdd.map(lambda row: row['Total']).max()
min_total = rdd.map(lambda row: row['Total']).min()

print(f"Max Total: {max_total}, Min Total: {min_total}")
```

```python
#FOREACH
# Utiliza foreach para aplicar una acción en cada elemento
rdd.foreach(lambda row: print(row))
# Contar por clave
city_counts = rdd.map(lambda row: (row["City"], 1)).countByKey()

# Usar foreach en el resultado para imprimir cada ciudad y su conteo
for city, count in city_counts.items():
    print(f"{city}: {count}")
```
